```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier

# Rótulos correspondentes às opções de reserva de hotel
rotulos = [
    "Reservar Quarto",
    "Modificar Reserva",
    "Cancelar Reserva",
    "Verificar Disponibilidade",
    "Ofertas Especiais e Promoções"
]

# Dados de treinamento
# Vetores de frases para cada opção de reserva de hotel
frases_reservar_quarto = [
    "Quero reservar um quarto para duas noites.",
    "Como faço para fazer uma reserva de quarto?",
    "Gostaria de reservar um quarto para o próximo fim de semana.",
    "Qual é o preço de uma suíte de luxo?",
    "Preciso de um quarto individual para três noites.",
]

frases_modificar_reserva = [
    "Quero mudar minha reserva para uma data diferente.",
    "Posso adiar minha reserva por uma semana?",
    "Modificar a data de check-in da minha reserva.",
    "Quero alterar o tipo de quarto da minha reserva.",
    "Adicionar uma noite à minha reserva existente.",
]

frases_cancelar_reserva = [
    "Preciso cancelar minha reserva.",
    "Não poderei mais viajar, então quero cancelar minha reserva.",
    "Como faço para cancelar minha reserva existente?",
    "Quero desfazer minha reserva para a próxima semana.",
    "Cancelar minha reserva com o número de confirmação 12345.",
]

frases_verificar_disponibilidade = [
    "Quais são as datas disponíveis para check-in?",
    "Posso fazer o check-in no próximo mês?",
    "Gostaria de verificar a disponibilidade para quartos de família.",
    "Quais quartos estão disponíveis para esta sexta-feira?",
    "Quero ver quartos com vista para o mar disponíveis.",
]

frases_ofertas_promocoes = [
    "Você tem alguma promoção atual?",
    "Mostre-me ofertas especiais para este mês.",
    "Gostaria de ver pacotes com desconto.",
    "Quais são as promoções de lua de mel?",
    "Existe algum desconto para estadias prolongadas?",
]

# Vetor principal contendo todas as categorias com rótulos e vetores de frases
mensagens_treinamento = frases_reservar_quarto + frases_modificar_reserva + frases_cancelar_reserva + frases_verificar_disponibilidade + frases_ofertas_promocoes

y = []
for rotulo in rotulos:
  for i in range(5):
    y.append(rotulo)

# Vetorização de texto usando TF-IDF
vectorizer = TfidfVectorizer()
x = vectorizer.fit_transform(mensagens_treinamento)

# Modelo KNN
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x, y)

# Função para classificar a entrada do usuário
def classificar_opcao(entrada_usuario):
    entrada_usuario_vetorizada = vectorizer.transform([entrada_usuario])
    classe_predita = knn.predict(entrada_usuario_vetorizada)
    return classe_predita[0]

# Exemplo de uso
entrada_do_usuario = input('Digite sua opção:')
classe_predita = classificar_opcao(entrada_do_usuario)
print(f"A entrada do usuário foi classificada como: {classe_predita}")
```
